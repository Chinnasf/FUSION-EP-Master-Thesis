```python
DB5.insert(loc = 0, column = "intercept", value = np.ones(len(DB5)))


y = DB5[["TAUTH"]].to_numpy()
x = DB5[["intercept"] + coeffs].to_numpy()
```

```python
np.matmul( np.linalg.inv( np.matmul(x.T, x) ) ,  np.matmul(x.T, y))
```
